```python
import fitparse
import pandas as pd
```

```python
fname = '18538612657_ACTIVITY'
```

```python
# load the FIT file
fitfile = fitparse.FitFile(f'./data/activities/runs/{fname}.fit')
```

```python
values = pd.DataFrame()
units = pd.DataFrame()

# each .fit file consists of various records
for record in fitfile.get_messages("record"):
    # each record has several data points, each with a name, a value, and (sometimes) a unit
    clms = [x.name for x in record]
    vlus = [x.value for x in record]
    unts = [x.units for x in record]

    temp_v = pd.DataFrame([vlus], columns = clms)
    temp_u = pd.DataFrame([unts], columns = clms)

    values = pd.concat([values, temp_v], ignore_index = True)
    units = pd.concat([units, temp_u], ignore_index = True)
```

```python
# save as csv
values.to_csv(f'./output/activities/runs/{fname}_values.csv', index = False)
units.to_csv(f'./output/activities/runs/{fname}_units.csv', index = False)
```
